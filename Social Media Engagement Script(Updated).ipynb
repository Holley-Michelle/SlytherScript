```python
import pandas as pd
from datetime import datetime
import collections
import itertools
import string
```

```python
# Read the CSV file
df = pd.read_csv('C:/Users/holle/OneDrive/Documents/Data Sets (Public)/Social Media Engagement Data.csv')
```

```python
#preview the dataframe
df.head()
```

```python
#Ensure the 'timestamp' column to datetime objects
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
#Extract date and time into new columns
df['Date'] = df['timestamp'].dt.date
df['Time'] = df['timestamp'].dt.time
```

```python
#ensure the new fields have been added they should be on the far right of the dataframe
df.head()
```

```python
# Split the 'location' column by comma and create new columns
df[['location_city', 'location_country']] = df['location'].str.split(',', expand=True)
```

```python
# Remove any leading/trailing whitespace from the new columns in case there was inconsistent spacing
df['location_city'] = df['location_city'].str.strip()
df['location_country'] = df['location_country'].str.strip()
```

```python
#preview change
df.head(2)
```

```python
# Function to split and clean hashtags
def clean_hashtags(text):
    if pd.isna(text):
        return []
    cleaned = []
    for h in text.split():
        # Remove leading #
        h = h.lstrip("#")
        # Remove trailing punctuation
        h = h.rstrip(string.punctuation)
        # Optional: convert to lowercase for consistency
        h = h.lower()
        cleaned.append(h)
    return cleaned
```

```python
# Function to split and clean keywords
def clean_keywords(text):
    if pd.isna(text):
        return []
    cleaned = []
    for k in text.split(","):
        k = k.strip()  # remove leading/trailing spaces
        k = k.rstrip(string.punctuation)  # remove trailing punctuation
        k = k.lower()  # optional: lowercase for uniformity
        cleaned.append(k)
    return cleaned
```

```python
# Function to split and clean mentions
def clean_mentions(text):
    if pd.isna(text):
        return []
    cleaned = []
    for m in text.split():
        m = m.lstrip("@")
        m = m.rstrip(string.punctuation)
        m = m.lower()
        cleaned.append(m)
    return cleaned
```

```python
# Apply functions to respective columns
hashtags = df['hashtags'].apply(clean_hashtags)
keywords = df['keywords'].apply(clean_keywords)
mentions = df['mentions'].apply(clean_mentions)
```

```python
# Flatten the lists
flat_hashtags = list(itertools.chain.from_iterable(hashtags))
flat_keywords = list(itertools.chain.from_iterable(keywords))
flat_mentions = list(itertools.chain.from_iterable(mentions))
```

```python
# Count using Counter
hashtag_counts = collections.Counter(flat_hashtags)
keyword_counts = collections.Counter(flat_keywords)
mention_counts = collections.Counter(flat_mentions)
```

```python
# Convert to DataFrames
hashtag_df = pd.DataFrame(hashtag_counts.items(), columns=["hashtag", "count"]).sort_values(by="count", ascending=False)
keyword_df = pd.DataFrame(keyword_counts.items(), columns=["keyword", "count"]).sort_values(by="count", ascending=False)
mention_df = pd.DataFrame(mention_counts.items(), columns=["mention", "count"]).sort_values(by="count", ascending=False)
```
